```python
balloons = [1, 3,1,5,8, 1]

mem = {}

def solve(arr):

    if tuple(arr) in mem:
        return mem[tuple(arr)]

    if len(arr[1:-1]) == 1:
        mem[tuple(arr)] = arr[1]
        return arr[1]


    mx = -1
    for i in range(1, len(arr) - 1):
        new_arr = arr.copy()
        new_arr.pop(i)
        sm = arr[i] * arr[i + 1] * arr[i - 1] + solve(new_arr)
        mx = max(mx, sm)
    mem[tuple(arr)] = mx

    return mem[tuple(arr)]

print(solve(balloons))



```
